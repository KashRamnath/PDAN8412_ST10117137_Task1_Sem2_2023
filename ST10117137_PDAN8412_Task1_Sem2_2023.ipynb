# What is RNN:
RNN, or Recurrent Neural Network, is a form of neural network. RNN functions through feeding the output of the prior step as an input to the current step in a sequential method, (Geeks for Geeks, 2023).

Recurrent neural networks utilize training data to learn. They are distinguished by their “memory” as they take information from prior inputs to influence the current input and output; which differs from regular deep learning neural networks which assumes inputs and outputs are independent from each other. These deep learning algorithms are commonly used for ordinal or temporal problems, such as language translation, natural language processing/NLP, speech recognition, and image captioning, (IBM, 2023).

RNN has the important feature known as memory state/hidden state. This allows it to remember data about a sequence and previous inputs to the network which reduces the intricacies of parameters opposed to other neural networks, (Geeks for Geeks, 2023).

RNNs have some disadvantages which include;

the training process is tricky, slow, and complex,
the issue with vanishing or exploding gradients (if at all the sequence gets too long, the model may train with null weights (i.e. no training) or exploding weights),
cannot be stacked due to its sequential input and output nature,
it is hard to process longer sequences.
(Open Genus, 2023).

# What is LSTM:
LSTM, or Long Short-Term Memory, is a sequential neural network in deep learning in which data is allowed to persist.

It’s a form of RNN which can handle the vanishing gradient problem faced by RNN as mentioned above. RNN can’t remember long-term dependencies due to vanishing gradient, but LSTMs are explicitly designed to avoid long-term dependency problems which makes it appropriate for sequence prediction applications (LSTM incorporates feedback connections which means it can process sequences of information and not just individual data points), (Analytics Vidhya, 2023).

LSTM has a slow training process due to its high intricacy,
but has higher performance than regular RNN over long sequences.
They are more complicated than traditional RNNs,
and require more training data to learn effectively.

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType, IntegerType
import nltk

# punkt is one of the modules in nltk.
# Punkt is made to learn parameters from a corpus in an unsupervised
# way that is related to the target domain, such as a list of
# abbreviations, acronyms, etc.
nltk.download('punkt')

@udf(returnType=StringType())
def clean_tweet(tweet):
# word_tokenize is a function in Python that splits a given sentence into words using the NLTK library.
  word_tokens = word_tokenize(tweet)

  filtered_sentence = [w.lower() for w in word_tokens if not w.lower() in stop_words and w.isalpha()]

  return ' '.join(filtered_sentence)
```

# Tokenization

```python
# Now that the data is split, we need to add
# tokenization, embedding, apdding,
# stemming and lemming.

# We start with tokenization.
# tokenization splits text into single words which assist
# in the training process.

# Tokenization is used in natural language processing to split paragraphs
# and sentences into smaller units that can be more easily assigned meaning

# first we import the keras tokenizer
from keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()

# Now that we activated the tokenizer, we
# fit it into the training data of the text column
tokenizer.fit_on_texts(train_data.clean_text)

# Each unique word is then assigned a unique integer index
word_index = tokenizer.word_index

# we now set a vocabulary size based on the index
vocab_size = len(tokenizer.word_index) + 1
print("Vocabulary Size :", vocab_size)
```

```python
# Label encoding can convert category labels such as positive,
# negative, and neutral into numerical representations in NLP
# applications such as text categorization or sentiment analysis.
# This enables machine learning models to comprehend and
# analyze text data successfully.
labels = train_data.target.unique().tolist()
```

```python
# lets begin label encoding
from sklearn import preprocessing

encoder = preprocessing.LabelEncoder()
encoder.fit(train_data.target.to_list())

# transform the data
y_train = encoder.transform(train_data.target.to_list())
y_test = encoder.transform(test_data.target.to_list())

y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)

print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

# Word Embedding

```python
# Reference: begin ---------------------------------------------
# GLOVE embedding adapted from Saturn Cloud
# Released: Monday, June 12, 2023
# Accessed: 15 September 2023
# URL: https://saturncloud.io/blog/how-to-use-glove-wordembeddings-file-on-google-colaboratory/#:~:text=You%20can%20download%20the%20GloVe,42B%2C%20and%20840B%20tokens).

# step 1 Install the Required Libraries
!pip install numpy gensim

# step 2 Download the GloVe Embeddings File
!wget http://nlp.stanford.edu/data/glove.6B.zip
!unzip glove.6B.zip -d glove.6B

#!wget http://nlp.stanford.edu/data/glove.6B.zip
#!unzip glove*.zip
```

```python
#GLOVE_EMB = '/kaggle/working/glove.6B.300d.txt'

# this link is to the files in
# googel colab when the previous
# cells downloads and unzips the glove folder
GLOVE_EMB = '/content/glove.6B/glove.6B.300d.txt'
EMBEDDING_DIM = 300
LR = 1e-3
BATCH_SIZE = 1024
EPOCHS = 10
MODEL_PATH = '.../output/kaggle/working/best_model.hdf5'
```

```python
# The advantage of GloVe is that, unlike Word2vec, GloVe does not
# rely just on local statistics (local context information of words),
# but incorporates global statistics (word co-occurrence)
# to obtain word vectors

embeddings_index = {}

f = open(GLOVE_EMB)
for line in f:
  values = line.split()
  word = value = values[0]
  coefs = np.asarray(values[1:], dtype='float32')
  embeddings_index[word] = coefs
f.close()

print('Found %s word vectors.' %len(embeddings_index))
```

```python
embedding_matrix = np.zeros((vocab_size, EMBEDDING_DIM))
for word, i in word_index.items():
  embedding_vector = embeddings_index.get(word)
  if embedding_vector is not None:
    embedding_matrix[i] = embedding_vector
```

```python
# Here the embedding layer is created

# The basic idea behind the GloVe word embedding is to
# derive the relationship between the words from statistics.

import tensorflow as tf

embedding_layer = tf.keras.layers.Embedding(vocab_size,
                                          EMBEDDING_DIM,
                                          weights=[embedding_matrix],
                                          input_length=MAX_SEQUENCE_LENGTH,
                                          trainable=False)
# Reference: end -----------------------------------------------------------------------
```

# LSTM

```python
# The dataframe is now ready for LSTM
# Lets start with gathering imports
from tensorflow.keras.layers import Conv1D, Bidirectional, LSTM, Dense, Input, Dropout
from tensorflow.keras.layers import SpatialDropout1D
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
sequence_input = Input(shape=(MAX_SEQUENCE_LENGTH,), dtype='int32')

# embedding layer - Generates Embedding Vector for each input sequence.
embedding_sequences = embedding_layer(sequence_input)

x = SpatialDropout1D(0.2)(embedding_sequences)

# to convolve data into smaller feature vectors
x = Conv1D(64, 5, activation='relu')(x)

x = Bidirectional(LSTM(64, dropout=0.2, recurrent_dropout=0.2))(x)

# Fully Connected Layers for classification
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)

# Fully Connected Layers for classification
x = Dense(512, activation='relu')(x)

# sigmoid maintains the values between 0 and 1.
# It helps the network to update or forget the data.
outputs = Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(sequence_input, outputs)
```

```python
# optimization algorithm for Gradient Descent.
# We will incorporate Adam.
# Adam(Adaptive Moment Estimation) is an adaptive
# optimization algorithm that was created specifically
# for deep neural network training.

# Why Adam?
# the Adam optimizer is considered to be effective
# for deep learning applications. It has been shown to
# perform well on a wide range of datasets and can help
# neural networks converge faster and more
# accurately during training,

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau

model.compile(optimizer=Adam(learning_rate=LR), loss='binary_crossentropy',
              metrics=['accuracy'])
ReduceLROnPlateau = ReduceLROnPlateau(factor=0.1,
                                     min_lr = 0.01,
                                     monitor = 'val_loss',
                                     verbose = 1)
```

```python
# Google colab has the option to run using a GPU
# however, if a GPU is not available, a CPU will
# suffice. Higher end GPUs are locked behind a paywall
# on Google Colab.
print("Training on GPU...") if tf.test.is_gpu_available() else print("Training on CPU...")
```

```python
# Here we will fit the model and print the process.
# Epoch is set to 10 with 1024 bach size
history = model.fit(x_train, y_train, batch_size=BATCH_SIZE, epochs=EPOCHS,
                    validation_data=(x_test, y_test), callbacks=[ReduceLROnPlateau])

# We will evaluate and visualise this history in
# the next step.
```

```python
# Here we have set 0.5 score because
# if the score is above above 0,5 then it will
# be classified as Positive sentiment.

def decode_sentiment(score):
    return "Positive" if score>0.5 else "Negative"


scores = model.predict(x_test, verbose=1, batch_size=10000)
y_pred_1d = [decode_sentiment(score) for score in scores]
```

```python
# confusion matrix

# imports
import itertools
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Plotting the data into confusion matrix
# for visualisation purposes
def plot_confusion_matrix(cm, classes,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    #"""
    #This function prints and plots the confusion matrix.
    #Normalization can be applied by setting `normalize=True`.
    #"""

    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, fontsize=20)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, fontsize=13)
    plt.yticks(tick_marks, classes, fontsize=13)

    fmt = '.2f'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label', fontsize=17)
    plt.xlabel('Predicted label', fontsize=17)
```

```python
# print the matrix
cnf_matrix = confusion_matrix(test_data.target.to_list(), y_pred_1d)
plt.figure(figsize=(6,6))
plot_confusion_matrix(cnf_matrix, classes=test_data.target.unique(), title="Confusion matrix")
plt.show()
```

```python
# classification scores
print(classification_report(list(test_data.target), y_pred_1d))
```
